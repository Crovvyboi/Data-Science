```python
import pandas as pd
import seaborn as sns

ytmpv = pd.read_csv('Youtube most popular videos US.csv', sep=',')
```

```python
ytmpv.describe()
```

```python
ytmpv.head()
```

```python
# What are the top 5 channels with the most amount of popular videos in this list?

df = pd.DataFrame(ytmpv)
sorted = df.groupby(['channel_title']).count().sort_values(["video_id"], ascending = False)
dropped = sorted.drop_duplicates(subset='title')
dropped["title"].head()
```

```python
# How many of the most popular videos have their comments disabled?

df = pd.DataFrame(ytmpv)
dropped = df.drop_duplicates(subset='title')
sorted = dropped.groupby(['comments_disabled']).count().sort_values(["video_id"])

sorted["title"]
```

```python
# What are, among the most popular videos, the top 5 videos with the worst like to dislike differences?

df = pd.DataFrame(ytmpv)

df['difference'] = df['likes'] - df['dislikes']

sorted = df.sort_values(['difference'])
dropped = sorted.drop_duplicates(subset='title').head()
dropped[['title', 'channel_title', 'difference']]
```

```python
# What is the average views per like on a popular video?

df = pd.DataFrame(ytmpv)

df = df[(df[['likes','views']] != 0).all(axis=1)]
df["viewsperlike"] = (df["views"]) / (df["likes"])

dropped = df.drop_duplicates(subset='title')
dropped["viewsperlike"].mean()
```
