```python
import pandas as pd
import seaborn as sns

ytmpv = pd.read_csv('Youtube most popular videos US.csv', sep=',')
```

```python
ytmpv.head()
```

Expectations:
    The strongest correlation is between views and likes.
    
Why?
    When people watch a video, they're more inclined to like a video than anything else, since there's a trend among youtubers to ask for likes before anything.

```python
ytmpv.plot(kind="scatter", x="views", y="likes")
```

The scatter plot is a positive logarithmic growth and comes close to a correlation between 0.6 and 0.9.
Some points (especially near the extremes) are nearly in a straight line, but dots near the center are a bit more scattered, which makes it a bit hard to tell.

```python
datacorr = ytmpv.corr()
datacorr.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

The correlation between views/likes is 0.85, which falls into range of the scatterchart estimation.

Both the scatter chart and Pearson matrix fall within my expectations, the Pearson matrix more than the scatterchart. I expected the correlation between views and likes to be high due to the assumption made before the analysis, which was confirmed by the numbers as highest of all correlations.
