```python
import pandas as pd
import seaborn as sns
def create_contingency_table(dataset, column1, column2):
    return dataset.groupby([column1, column2]).size().unstack(column1, fill_value=0)


ytmpv = pd.read_csv('Youtube most popular videos US.csv', sep=',')
```

```python
ytmpv.head()
```

Is there any correlation between the category and comments_disabled?

Expectations
 - I expect there to be some correlation between disabled comments and category. I don't know which category it will be (because there are no names, only id's), but some categories, like news stories handling sensitive topics, are more likely to have their comments disabled.
 - I also expect a higher comment disabled count the more videos a category has.

```python
contTable = create_contingency_table(ytmpv, 'category_id','comments_disabled')
contTable
```

```python
contTable.plot(kind='bar')
```

From the table, expectation 2 can be seen as false. Even though categories like 24, 28 and 22 have more often their comments disabled and more videos in the category (sum from both rows), there are also categories, often with a similar or more videos in the category, that have less times their comments disabled, like category 10, 23 and 26.

This, however, lends credence to expectation 1. Since there's no correlation between videos in a category and the frequency of their comments being disabled, the fact that the category itself has an influence on if the comments are more often disabled might be more true than not. I cannot definitively name the reason for this being, but I still suspect the content or crowd that the content attracts has something to do with it.
