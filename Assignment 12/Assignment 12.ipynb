```python
import pandas as pd
import seaborn as sns

ytmpv = pd.read_csv('Youtube most popular videos US.csv', sep=',')
```

```python
ytmpv.head()
```

```python
ytmpv.groupby("category_id").mean()
```

There are significant differences in view means, like means, dislike means and comment_count means

For the sake of this assignment, I will pick views, likes and comment_count to further explore

```python
sns.barplot(y="views", x="category_id", data=ytmpv)
```

In category 10 the views are the most and notably with a great confidence. 

Also notable is the views of category 29, which either has (relatively speaking) a lot of views or very little.

All other categories' means are pretty accurate to the confidence.

```python
sns.barplot(y="likes", x="category_id", data=ytmpv)
```

Like the views, the likes of category 29 fluctuate a lot, not giving the means much confidence.

The means of category 10 (and all others) are a bit more confident.

```python
sns.barplot(y="comment_count", x="category_id", data=ytmpv)
```

There's much more engagement in category 29, even with the shaky confidence on the means.

While the confidence of other categories is more accurate to the means, they do not come close to the amount of comments in category 29
